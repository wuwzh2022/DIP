```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./figure/Nuts.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
thresh = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY)[1]
plt.imshow(thresh, cmap='gray')
plt.axis('off')
plt.show()
```
